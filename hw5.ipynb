```python
from xenonnt_plot_style import XENONPlotStyle as xps
xps.use('xenonnt')
```

```python
import numpy as np
import matplotlib.pyplot as plt

from utils import k0, beta
from utils import tm_amplitude
from plot import plot_kf, plot_V
```

## 1

```python
lam = 0.8
```

```python
nf = 1.5
ns = 1.495
nc = 1.0
h = 2.0
# tm_root = plot_kf(lam, nf, ns, nc, h, te=False)[0]
# neff1 = beta(tm_root, lam, nf) / k0(lam)
root = plot_V(lam, nf, ns, nc, h, te=False, numax=3)[0]
neff1 = (root * (nf ** 2 - ns ** 2) + ns ** 2) ** 0.5
```

```python
# H_y of first TM mode

x = np.linspace(-2 * h, h, 10001)

fig, ax = plt.subplots(1, 1, figsize=(4, 3))

kf = (nf ** 2 - neff1 ** 2) ** 0.5 * k0(lam)
ax.plot(x, tm_amplitude(x, kf, lam, nf, ns, nc, h))
ax.axvline(0, linestyle='dashed', color=xps.colors['grey'])
ax.axvline(-h, linestyle='dashed', color=xps.colors['grey'])

ax.set_xlim(x[0], x[-1])

ax.set_xlabel('x [$\mu$m]')
ax.set_ylabel('$H_y$')

plt.show()
```

```python
nf = neff1
ns = 1.495
nc = 1.495
h = 10.0
# tm_root = plot_kf(lam, nf, ns, nc, h, te=False)[0]
# neff = beta(tm_root, lam, nf) / k0(lam)
root = plot_V(lam, nf, ns, nc, h, te=False, numax=3)[0]
neff = (root * (nf ** 2 - ns ** 2) + ns ** 2) ** 0.5
```

```python
# H_y of first TM mode

x = np.linspace(-2 * h, h, 10001)

fig, ax = plt.subplots(1, 1, figsize=(4, 3))

kf = (nf ** 2 - neff ** 2) ** 0.5 * k0(lam)
ax.plot(x, tm_amplitude(x, kf, lam, nf, ns, nc, h))
ax.axvline(0, linestyle='dashed', color=xps.colors['grey'])
ax.axvline(-h, linestyle='dashed', color=xps.colors['grey'])

ax.set_xlim(x[0], x[-1])

ax.set_xlabel('x [$\mu$m]')
ax.set_ylabel('$H_y$')

plt.show()
```
